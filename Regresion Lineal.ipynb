```python
x = regrelineal[['precio']]
y = regrelineal[['metro']]
```

```python
model = LinearRegression()
model.fit(x, y)

predi_reg_lineal = model.predict(x)

plt.scatter(x, y, color='grey')
plt.plot(x, predi_reg_lineal, color='black')
plt.xlabel('Precio')
plt.ylabel('metro')
plt.title('Grafica de Regresión lineal precio y metro')
plt.show()
```
