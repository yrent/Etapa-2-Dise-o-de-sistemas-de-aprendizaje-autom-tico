```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.tree import plot_tree
```

```python
nombres = ['Class', 'Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash', 'Magnesium', 'Total phenols', 'Flavanoids', 'Nonflavanoid phenols', 'Proanthocyanins', 'Color intensity', 'Hue', 'OD280/OD315 of diluted wines', 'Proline']
df = pd.read_csv("C:/Users/hp/OneDrive/Escritorio/PERIODO 16 02 2023/ANALISIS DE DATOS/ETAPA 2/wine.data", names=nombres)
```

```python
predictors_col = ["Class", "Alcohol", "Malic acid", "Ash", "Alcalinity of ash", "Magnesium", "Total phenols", "Flavanoids", "Nonflavanoid phenols", "Proanthocyanins", "Color intensity", "Hue", "OD280/OD315 of diluted wines"]
target_col = ['Proline']
```

```python
predictors = df[predictors_col]
target = df[target_col]
```

```python
X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.2, random_state=42)
```

```python
tree = DecisionTreeClassifier()
```

```python
arbol=tree.fit(X_train, y_train)
```

```python

plot_tree(arbol)
```

```python
predicciones = arbol.predict(X_test)
```

```python
pd.crosstab(np.array([y[0] for y in y_test.values.tolist()]), predicciones, rownames=['Actual'], colnames=['Predicciones'])
```

```python
accuracy = accuracy_score(y_test, predicciones)
accuracy
```
